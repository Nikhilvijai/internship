```python
from sklearn.preprocessing import MinMaxScaler,MaxAbsScaler,StandardScaler,RobustScaler

file_path = '/content/titanic uncleaned.csv'
titanic_data = pd.read_csv(file_path)
```

```python
numerical_columns = ['Age','Fare','Pclass']
```

```python
titanic_data.dropna(subset = numerical_columns,inplace=True)

minmax_scalar = MinMaxScaler()
standard_scalar = StandardScaler()
```

```python
titanic_data['Age_MinMax'] = minmax_scalar.fit_transform(titanic_data[['Age']])
titanic_data['Fare_MinMax'] = minmax_scalar.fit_transform(titanic_data[['Fare']])

titanic_data['Age_standard'] = standard_scalar.fit_transform(titanic_data[['Age']])
titanic_data['Fare_standard'] = standard_scalar.fit_transform(titanic_data[['Fare']])
```
